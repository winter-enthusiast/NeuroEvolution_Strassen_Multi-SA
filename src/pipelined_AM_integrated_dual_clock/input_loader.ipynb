```python
import numpy as np

idx = [
    [0, 4, 8, 12],
    [1, 5, 9, 13],
    [2, 6, 10, 14],
    [3, 7, 11, 15],
]

A = np.array([
    [3, 4, 2, 4, 4, 1, 2, 2, 2, 4, 3, 2, 5, 4, 1, 3],
    [5, 5, 1, 3, 4, 0, 3, 1, 5, 4, 3, 0, 0, 2, 2, 1],
    [3, 3, 5, 5, 5, 2, 3, 3, 0, 2, 4, 2, 4, 0, 1, 3],
    [0, 3, 5, 1, 1, 0, 1, 4, 1, 3, 3, 3, 3, 4, 2, 5],
    [0, 3, 1, 3, 1, 5, 5, 5, 1, 3, 5, 4, 1, 1, 3, 1],
    [1, 5, 3, 5, 5, 3, 0, 5, 4, 4, 1, 4, 1, 0, 3, 3],
    [3, 4, 0, 4, 4, 0, 0, 0, 0, 3, 2, 2, 0, 2, 2, 0],
    [2, 4, 1, 1, 0, 3, 0, 3, 1, 0, 5, 4, 2, 3, 5, 2],
    [2, 0, 2, 4, 5, 2, 0, 4, 1, 5, 2, 0, 1, 1, 3, 4],
    [2, 0, 3, 4, 3, 5, 4, 4, 2, 4, 3, 4, 2, 2, 5, 3],
    [1, 1, 4, 5, 0, 4, 5, 3, 3, 3, 3, 3, 5, 5, 2, 1],
    [3, 0, 5, 0, 0, 0, 2, 5, 0, 3, 4, 0, 2, 5, 2, 0],
    [5, 4, 0, 2, 1, 3, 2, 5, 0, 3, 0, 5, 0, 1, 3, 3],
    [5, 1, 2, 0, 4, 0, 0, 2, 0, 1, 1, 3, 5, 4, 0, 0],
    [2, 5, 1, 4, 3, 1, 5, 3, 2, 2, 0, 5, 4, 3, 1, 5],
    [5, 2, 0, 0, 3, 2, 5, 4, 2, 3, 3, 2, 3, 2, 1, 2],
])

B = np.array([
    [2, 3, 3, 0, 0, 1, 0, 2, 3, 0, 0, 1, 5, 1, 2, 3],
    [1, 0, 3, 3, 0, 1, 0, 5, 3, 4, 4, 2, 0, 0, 2, 2],
    [2, 3, 5, 0, 3, 2, 0, 3, 5, 3, 5, 2, 5, 0, 2, 0],
    [4, 1, 5, 1, 1, 5, 2, 4, 0, 3, 0, 3, 0, 5, 4, 3],
    [2, 0, 0, 3, 2, 2, 5, 5, 5, 5, 5, 5, 4, 2, 5, 2],
    [2, 1, 4, 5, 0, 3, 0, 4, 3, 4, 2, 3, 2, 0, 0, 3],
    [3, 5, 4, 5, 4, 5, 2, 3, 0, 4, 4, 0, 5, 4, 2, 3],
    [0, 3, 4, 4, 0, 2, 1, 0, 1, 5, 1, 2, 5, 1, 5, 1],
    [2, 1, 1, 1, 0, 0, 0, 2, 5, 4, 1, 1, 2, 1, 0, 4],
    [3, 1, 0, 3, 4, 3, 0, 5, 5, 3, 2, 3, 1, 1, 5, 2],
    [0, 1, 5, 4, 5, 1, 1, 5, 0, 3, 1, 5, 5, 2, 3, 4],
    [0, 4, 3, 3, 3, 4, 3, 5, 4, 3, 5, 2, 3, 4, 1, 3],
    [1, 5, 2, 0, 2, 3, 1, 1, 4, 1, 4, 5, 0, 3, 4, 0],
    [1, 1, 0, 1, 5, 5, 0, 4, 4, 0, 4, 4, 5, 4, 2, 3],
    [1, 2, 4, 5, 0, 4, 5, 3, 4, 0, 5, 5, 5, 3, 4, 3],
    [1, 1, 4, 3, 0, 5, 5, 4, 1, 1, 5, 4, 3, 1, 3, 1],
])

def generate_coe_from_rows_concatenated(filename, array, row_indices_list):
    with open(filename, 'w') as f:
        f.write("memory_initialization_radix=16;\n")
        f.write("memory_initialization_vector=\n")
        for row_indices in row_indices_list:
            concatenated_binary = ""
            for row_index in row_indices:
                if 0 <= row_index < array.shape[0]:
                    for element in array[row_index]:
                        binary_representation = format(element, '08b')
                        concatenated_binary += binary_representation
                else:
                    print(f"Warning: Row index {row_index} is out of bounds for array of shape {array.shape}.")

            if concatenated_binary:
                hex_representation = format(int(concatenated_binary, 2), '0' + str(len(concatenated_binary) // 4) + 'x')
                f.write(hex_representation + ",\n")
            else:
                f.write("0,\n") # Write 0 if no data was processed
        f.write(";\n")

def generate_coe_from_cols_concatenated(filename, array, col_indices_list):
    with open(filename, 'w') as f:
        f.write("memory_initialization_radix=16;\n")
        f.write("memory_initialization_vector=\n")
        for col_indices in col_indices_list:
            concatenated_binary = ""
            for col_index in col_indices:
                if 0 <= col_index < array.shape[1]:
                    for element in array[:, col_index]:
                        binary_representation = format(element, '08b')
                        concatenated_binary += binary_representation
                else:
                    print(f"Warning: Column index {col_index} is out of bounds for array of shape {array.shape}.")

            if concatenated_binary:
                hex_representation = format(int(concatenated_binary, 2), '0' + str(len(concatenated_binary) // 4) + 'x')

                # THis is because in STRASSEN RTL we are accesing in a reverse way
                new_hex_2letter_reversal = ''
                for i in range(0, len(hex_representation), 2):
                    new_hex_2letter_reversal = hex_representation[i:i+2] + new_hex_2letter_reversal
                print(new_hex_2letter_reversal)

                f.write(new_hex_2letter_reversal + ",\n")
            else:
                f.write("0,\n") # Write 0 if no data was processed
        f.write(";\n")

# Generate bram_A.coe
bram_A_row_indices = idx
generate_coe_from_rows_concatenated("bram_A.coe", A, bram_A_row_indices)
print("bram_A.coe generated successfully.")

# Generate bram_B.coe
bram_B_col_indices = idx
generate_coe_from_cols_concatenated("bram_B.coe", B, bram_B_col_indices)
print("bram_B.coe generated successfully.")
```
