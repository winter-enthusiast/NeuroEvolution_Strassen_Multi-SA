```python
import numpy as np

# Generate random 16x16 matrices with integers between 0 and 9
np.random.seed(42)  # Set seed for reproducibility
A = np.random.randint(0, 10, (16, 16))
B = np.random.randint(0, 10, (16, 16))


# T1 of r=1
A = np.array(
    [[14,  7 ,12 , 7 ,15 ,15 ,10 ,12],
    [ 6, 14 ,13 , 5 , 4 , 2 ,18 ,12],
    [10,  3 , 8 , 4 ,12 , 6 , 8 , 3],
    [ 8,  8 , 6 , 9 , 8 , 7 ,10 , 9],
    [ 6,  9 ,11 ,12 , 8 ,11 ,10 , 6],
    [13, 14 , 8 , 7 , 7 ,15 ,10 , 1],
    [ 9,  6 ,12 ,11 , 7 , 8 ,14 , 8],
    [ 5,  2 , 4 , 8 ,15 ,14 ,14 ,10]],
)

# S1 of r = 1
B = np.array(
    [[14,  7, 12 , 7 ,15, 15, 10, 12,],
    [ 6 , 14, 13 , 5 , 4,  2, 18, 12,],
    [10 ,  3,  8 , 4 ,12,  6,  8,  3,],
    [ 8 ,  8,  6 , 9 , 8,  7, 10,  9,],
    [ 6 ,  9, 11 ,12 , 8, 11, 10,  6,],
    [13 , 14,  8 , 7 , 7, 15, 10,  1,],
    [15 , 12,  9 ,14 , 1,  8, 14,  7,],
    [ 5 ,  2,  4 , 8 ,15, 14, 14, 10,],]
)



print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
```

```python
def split_matrix(A):
    """Splits a square matrix A into four equal quadrants."""
    n = A.shape[0] // 2
    return A[:n, :n], A[:n, n:], A[n:, :n], A[n:, n:]

# Split matrices into quadrants
A11, A12, A21, A22 = split_matrix(A)
B11, B12, B21, B22 = split_matrix(B)

# Print quadrants
print("Quadrants of A:")
print("A11:\n", A11, "\nA12:\n", A12, "\nA21:\n", A21, "\nA22:\n", A22)

print("\nQuadrants of B:")
print("B11:\n", B11, "\nB12:\n", B12, "\nB21:\n", B21, "\nB22:\n", B22)
```

```python
# Compute T matrices
T1 = A11 + A22
T2 = A21 + A22
T3 = A11
T4 = A22
T5 = A11 + A12
T6 = A21 - A11
T7 = A12 - A22

# Compute S matrices
S1 = B11 + B22
S2 = B11
S3 = B12 - B22
S4 = B21 - B11
S5 = B22
S6 = B11 + B12
S7 = B21 + B22

# Print T matrices
print("T Matrices:")
for i, T in enumerate([T1, T2, T3, T4, T5, T6, T7], 1):
    print(f"T{i}:\n{T}\n")

# Print S matrices
print("S Matrices:")
for i, S in enumerate([S1, S2, S3, S4, S5, S6, S7], 1):
    print(f"S{i}:\n{S}\n")
```

```python
# Compute Q matrices
Q1 = T1 @ S1
Q2 = T2 @ S2
Q3 = T3 @ S3
Q4 = T4 @ S4
Q5 = T5 @ S5
Q6 = T6 @ S6
Q7 = T7 @ S7

# Print Q matrices
print("Q Matrices:")
for i, Q in enumerate([Q1, Q2, Q3, Q4, Q5, Q6, Q7], 1):
    print(f"Q{i}:\n{Q}\n")
```

```python
# Compute C submatrices
C11 = Q1 + Q4 - Q5 + Q7
C12 = Q3 + Q5
C21 = Q2 + Q4
C22 = Q1 - Q2 + Q3 + Q6

# Reconstruct full C matrix
C_top = np.hstack((C11, C12))
C_bottom = np.hstack((C21, C22))
C = np.vstack((C_top, C_bottom))

print("Final Matrix C = A * B:")
print(C)
```

```python

# Define input tensor (C_in=3, H=3, W=4)
input_tensor = np.random.rand(3, 5, 4)  # 3 channels, 3x4 image

# Define kernel (C_in=3, H=2, W=2, C_out=1)
kernel = np.random.rand(1, 3, 2, 2)  # 1 output channel, 3 input channels

# Traditional convolution
output_height = input_tensor.shape[1] - kernel.shape[2] + 1
output_width = input_tensor.shape[2] - kernel.shape[3] + 1
output_traditional = np.zeros((1, output_height, output_width))

for oc in range(1):  # Output channels
    for h in range(output_height):
        for w in range(output_width):
            output_traditional[oc, h, w] = np.sum(
                input_tensor[:, h:h+2, w:w+2] * kernel[oc]
            )

# Convolution using GEMM

# Step 1: Unfold input into a matrix
patches = []
for h in range(output_height):
    for w in range(output_width):
        patches.append(input_tensor[:, h:h+2, w:w+2].reshape(-1))
input_matrix = np.array(patches)  # Shape: (output_height * output_width, C_in * kernel_h * kernel_w)

# Step 2: Reshape kernel
kernel_matrix = kernel.reshape(1, -1)  # Shape: (C_out, C_in * kernel_h * kernel_w)

# Step 3: Compute GEMM-based convolution
output_gemm = (input_matrix @ kernel_matrix.T).T.reshape(1, output_height, output_width)

# Compare results
print("Traditional Convolution Result:")
print(output_traditional)
print("\nGEMM-based Convolution Result:")
print(output_gemm)
print("\nDifference:")
print(output_traditional - output_gemm)
```
