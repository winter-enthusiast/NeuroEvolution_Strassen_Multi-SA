```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generate_matrix(size, low=0, high=5):
    return np.random.randint(low, high + 1, (size, size))

np.random.seed(42)  # For reproducibility
A = generate_matrix(16)
B = generate_matrix(16)

# Create a figure with three subplots
fig, axes = plt.subplots(3, 1, figsize=(6, 12))

# Plot matrix A
sns.heatmap(A, annot=True, fmt="d", cmap="viridis", ax=axes[0])
axes[0].set_title("Matrix A")

# Plot matrix B
sns.heatmap(B, annot=True, fmt="d", cmap="plasma", ax=axes[1])
axes[1].set_title("Matrix B")

# Compute matrix C as the product of A and B
C = A @ B

# Plot matrix C
sns.heatmap(C, annot=True, fmt="d", cmap="cubehelix", ax=axes[2])
axes[2].set_title("Matrix C")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Function to plot matrices in a grid
def plot_matrices(matrices, title, cmap, rows, cols=3):
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 4))
    fig.suptitle(title, fontsize=16)
    axes = axes.ravel()  # Flatten the axes array for easy iteration
    for i, matrix in enumerate(matrices):
        sns.heatmap(matrix, annot=True, fmt="d", cmap=cmap, ax=axes[i], cbar=False)
        axes[i].set_title(f"{title[:-1]} {i+1}")
    # Hide unused subplots
    for j in range(len(matrices), rows * cols):
        axes[j].axis("off")
    plt.tight_layout()
    plt.show()

# Plot T-matrices (7 matrices, 3 per row)
plot_matrices(T_matrices, "T-matrices", cmap="viridis", rows=3)

# Plot S-matrices (7 matrices, 3 per row)
plot_matrices(S_matrices, "S-matrices", cmap="plasma", rows=3)
```
